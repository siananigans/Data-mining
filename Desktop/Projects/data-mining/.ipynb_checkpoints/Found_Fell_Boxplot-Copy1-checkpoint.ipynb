```python
fall = data['fall']
```

```python
fall.tolist()
```

```python
fell = []
found = []
i = 0
while i < len(fall):
    if fall[i] == 'Fell':
        fell.append(lat[i])
    elif fall[i] == 'Found':
        found.append(lat[i])
        
    i+=1
    
print(len(fell) + len(found))
```

```python
lat_found = []
for o in found:
    if pandas.notnull(o):
        lat_found.append(o)
        
print(lat_found)
```

```python
lat_fell = []
for o in fell:
    if pandas.notnull(o):
        lat_fell.append(o)
        
print(lat_fell)
```

```python
def boxplot(a):
    print(matplotlib.pyplot.boxplot(a))
    
boxplot(lat_fell)
```

```python

lat = [lat_found, lat_fell]
labels = ("Found:Lat", "Fell:Lat")

print(lat)

matplotlib.pyplot.boxplot(lat,labels=labels)
```
